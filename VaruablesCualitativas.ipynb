```python
 # Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
nivel_estudios = []
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)
nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
 # DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
 # DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Recreamos los datos originales del ejercicio
nivel_estudios = []
nivel_estudios.extend(['Secundaria'] * 110)
nivel_estudios.extend(['Bachillerato'] * 172)
nivel_estudios.extend(['Grado'] * 145)
nivel_estudios.extend(['Master'] * 52)
nivel_estudios.extend(['Doctorado'] * 21)
```

```python
# Convertimos la lista tradicional de Python a una Serie de Pandas
df_estudios = pd.Series(nivel_estudios)
```

```python
# 2. Contar frecuencias con Pandas (El equivalente a np.unique)
# .value_counts() cuenta automáticamente cuántas veces aparece cada categoría
frecuencias = df_estudios.value_counts()
```

```python
print("--- Datos procesados con Pandas ---")
print(frecuencias)
```

```python
# 3. DIAGRAMA DE BARRAS CON PANDAS
# En lugar de usar plt.bar(), le decimos a nuestra serie que se dibuje a sí misma
frecuencias.plot(kind='bar', figsize=(8, 5))
plt.title('Diagrama de barras')
plt.xlabel('Nivel de estudios')
plt.ylabel('Frecuencia absoluta')
plt.xticks(rotation=0) # Mantiene los nombres horizontales para que sean fáciles de leer
plt.show()
```

```python
# 4. DIAGRAMA DE SECTORES CON PANDAS
# Usamos kind='pie' y le pasamos los mismos parámetros de porcentaje
colores = ['orange', 'blue', 'red', 'cyan', 'green']
frecuencias.plot(kind='pie', autopct='%1.2f%%', colors=colores, figsize=(7, 7))
plt.title('Diagrama de sectores')
plt.ylabel('') # Oculta el nombre del eje 'y' que Pandas pone por defecto en los pasteles
plt.show()
```
